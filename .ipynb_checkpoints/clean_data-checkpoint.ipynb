```python
import pandas as pd
```

```python
apr = './Resources/Apr-2021.csv'
feb = './Resources/Feb-2021.csv'
jan = './Resources/Jan-2021.csv'
may = './Resources/May-2021.csv'
```

```python
apr_df = pd.read_csv(apr)
apr_df.head()
```

```python
feb_df = pd.read_csv(feb)
feb_df.head()
```

```python
jan_df = pd.read_csv(jan)
jan_df.head()
```

```python
may_df = pd.read_csv(may)
may_df.head()
```

```python
combined_df = jan_df.merge(feb_df, how="outer")
combined_df.head()
```

```python
combined_df = combined_df.merge(apr_df, how="outer")
combined_df.head()
```

```python
combined_df = combined_df.merge(may_df, how="outer")
combined_df.head()
```

```python
combined_df = combined_df.dropna()
```

```python
combined_df.to_csv('./Resources/clean_data.csv')
```
